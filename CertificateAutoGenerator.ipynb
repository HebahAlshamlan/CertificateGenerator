# Arabic text

## Import

```
# use a truetype font  
font = ImageFont.truetype("Tajawal-Regular.ttf", 100)  
```

```
dataset = pd.read_excel('Book1.xlsx')
dataset
```

```
names = dataset['Unnamed: 0'].copy()
hours = dataset.drop(['Unnamed: 0'],axis=1).copy()
```

```
n = hours.iloc[1]
n[0]
```

```
for i in range(dataset.shape[0]):
    image = Image.open("شهادة متطوعات كسوة فرح ٧.jpg")  
    draw = ImageDraw.Draw(image)
    VulName = names.iloc[i]
    hour = hours.iloc[i]
    reshaped_text = arabic_reshaper.reshape(VulName) # Reshap Arabic text
    rev_text = reshaped_text[::-1]  # slice backwards 
    draw.text((1500 , 970 ), rev_text, fill=(255,0,0,255), font=font)   # First Col
    draw.text((1810 , 1255 ), str(hour[0]), font=font)  # second col
    plt.imshow(image)
    im1 = image.convert('RGB') #Convert to save as PDF
    im1.save(r'Ms.'+names.iloc[i]+'.pdf') # Save as a PDF
    # image.save("no"+str(i)+".jpg")  
```
